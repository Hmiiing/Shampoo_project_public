## 유니코드를 한글로 변환하는 함수

```python
## 유니코드를 한글로 변환하는 함수

def uni2han(uni) :
    # "유니코드->한글 입력창"에 "유니코드" 넣기
    input_ = driver.find_element_by_css_selector("#input")
    driver.find_element_by_css_selector("#input").click()

    # 유니코드 입력
    input_.clear()
    input_.send_keys(uni)

    # URL 디코더 버튼 누르기
    driver.find_element_by_css_selector('#submit').click()    

    # 한글 가져오기
    output_ = driver.find_element_by_css_selector("#output").text
    out = output_.replace('<br>', '').replace('\n', ' ').replace('<p>',' ').replace('</p>','')
    
    return out
```
