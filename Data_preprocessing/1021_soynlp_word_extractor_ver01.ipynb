```python
import sys
sys.path.insert(0, '../')
from soynlp.utils import DoublespaceLineCorpus
    
#corpus_fname = 'mycorpus.txt' #이 코퍼스로 하는거 아님! 
corpus_fname = 'mycorpus_줄바꿈_적용.txt'   # 이걸로 하세요 여러분 
sentences = DoublespaceLineCorpus(corpus_fname, iter_sent=True)
print('num sentences = %d' % len(sentences))
```

### train은 substrings의 빈도수를 카운팅 하는 것이며, extract는 init에 들어가는 값을 기준으로 단어를 선택하여 줍니다.

```python
%%time

from soynlp.word import WordExtractor

word_extractor = WordExtractor(
    min_frequency=5,
    min_cohesion_forward=0.05, 
    min_right_branching_entropy=0.0
)

word_extractor.train(sentences)
words = word_extractor.extract()
```

```python
len(words)
```

```python
words
```

```python
result = []
for key in words:
    if 2 <= len(key):
        result.append(key)
         
print(result)
```

## 코퍼스 줄 단위로 안끊어져있는거로 할 때  ; 코퍼스 txt '줄바꿈 적용'이라고  써있는걸로 하세요~
 
- 코퍼스가 리뷰 줄단위로 구분 안되어있는걸로 할 경우, a라는 리뷰의 끝어절과 a+1리뷰의 첫 어절이 붙어버리는 경우가 있어서 토큰화가 잘 안되는 문제가 있습니다. 그래서 리뷰 하나하나 개행문자 적용(= 줄바꿈 적용)한 걸로 단어를 추출하시기 바랍니다.

#### 아래 결과는 코퍼스 줄 단위 안되어있는걸로 했을 때의 결과입니다. 뒤에 뽑힌거 보면 리뷰 끝어절과 그다음 리뷰 첫어절이 하나의 어절로 뽑힌거 같은게 보입니다! 

```python
result_2 = []
for word in result:
    if '향' in word:
        result_2.append(word)
        
print(result_2)
```
