```python
import pandas as pd

df = pd.read_excel('excel_02_1018_1단계_CJ_review_peprocessing.xlsx', encoding='cp949')
```

```python
df.head()
```

```python
df_review = df['review']
```

```python
mystr = ''

for i in range(0, len(df_review), 1):
    mystr += str(df_review[i]) + '\n'
```

```python
mystr[:1000]
```

```python
len(mystr)
```

### 파일 이름은 원하는대로~

```python
with open('mycorpus_example.txt', 'w', encoding='utf-8') as f:
    f.write(mystr)
```
