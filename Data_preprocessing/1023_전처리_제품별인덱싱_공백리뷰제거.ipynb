#### 공백값이 다 길이가 3 인지 한 번 확인해보자!

```python
#문자길이 칼럼 생성
허벌['문자길이'] = 0
```

```python
허벌.head(5)
```

```python
for i in range(len(허벌)): 
    허벌['문자길이'].iloc[i] = len(허벌['review'].iloc[i])
```

```python
허벌.head(50) # 크롤링 된 곳에서 리뷰 안의 공백값 길이는 3인 것 같다. 따라서 위에서 사용된 공백 길이로 불린 인덱싱하자! 
             # 19,20번째 로우 보면 review 공백이다. 여기서도 길이는 3. 
             # 아무래도 크롤링될 때 공백 부분은 space 길이가 3인 공백으로 들어온 거 같다. 
```

```python
space_erase01_허벌.shape # 11개의 공백을 제거!
```

#### reset_index

```python
공백제거허벌에센스_resetindex = space_erase01_허벌.reset_index(inplace=False)
```

```python
공백제거허벌에센스_resetindex.iloc[10:60]
```

#### reset_index를 하면 기존 index가 Index 칼럼으로 새로 생성되며 처음부터 indexing이 된다. 따라서 기존 index 칼럼은 삭제해줘야 한다.

```python
공백제거허벌에센스_resetindex.drop(['index'],axis='columns' , inplace = True)
```

```python
공백제거허벌에센스_resetindex.iloc[10:60]
```
