```python

user_list=[]
star_list=[]  
review_list=[]
brand_name=[]
product_name=[]

driver = webdriver.Chrome("chromedriver.exe")     
#피부타입 클릭
for i in range(len(mid_info_url)):
    driver.get(o_url['url'][i+377]) # 나중에 69부터 100까지 다시하기

    for kk in range(5):
        sample1 = driver.find_element_by_css_selector('.fieldset:nth-child(3) li:nth-child('+str(kk+2)+')')
        driver.execute_script("arguments[0].click();", sample1)
        time.sleep(0.3)
        sample2 = driver.find_element_by_css_selector('.filter-submit__submit-button')
        driver.execute_script("arguments[0].click();", sample2) 
        time.sleep(0.5)
        sample1 = driver.find_element_by_css_selector('.fieldset:nth-child(3) li:nth-child('+str(kk+2)+')')
        driver.execute_script("arguments[0].click();", sample1)
        time.sleep(2)

        l=0
        if int(driver.find_element_by_css_selector('.section-list-score__rating-item.joiner').text.replace('총 ',"").replace('명',"")) >= 5:
            while l<5:
                # 스크롤 제일 마지막 아래 대상 선택
                ELEMENT = driver.find_elements_by_css_selector('.review-list-item')[-2]
                time.sleep(0.5)

                # ELEMENT 가 화면에 보이도록 스크롤 조정 --> 아래 부분 추가로 확장됨
                driver.execute_script("arguments[0].scrollIntoView(true);", ELEMENT)        
                l=l+1
                time.sleep(0.5)

                # 스크롤 제일 마지막 아래 대상 선택
                ELEMENT = driver.find_elements_by_css_selector('.review-list-item')[-5]
                time.sleep(1)

                # ELEMENT 가 화면에 보이도록 스크롤 조정 --> 아래 부분 추가로 확장됨
                driver.execute_script("arguments[0].scrollIntoView(true);", ELEMENT)                

        else:                 
            time.sleep(0.5)                


        #아이디  user_list=[]
        users = driver.find_elements_by_css_selector(".review-list-item .user-name")
        for j in users:
            user_list.append(j.text)
        user_list



        #별 star_list=[]  
        stars = driver.find_elements_by_css_selector(".review-list-item .label span")
        for v in stars:
            if v.get_attribute('class') == 'icon-sprite rating-grade-icon gpa-best-small':
                star_list.append('별다섯개')
            elif v.get_attribute('class') == 'icon-sprite rating-grade-icon gpa-good-small':
                star_list.append('별네개')
            elif v.get_attribute('class') == 'icon-sprite rating-grade-icon gpa-soso-small':
                star_list.append('별세개')            
            elif v.get_attribute('class') == 'icon-sprite rating-grade-icon gpa-bad-small':
                star_list.append('별두개')                    
            else:
                star_list.append('별한개')

        # 리뷰 크롤링     review_list=[]
        overlays1 = ".review-list-item p.review"                     
        reviews = driver.find_elements_by_css_selector(overlays1)        
        for k in reviews:
            review_remove=review_list.append(k.text.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","").replace('\n'," "))
            #브랜드 이름
            s=driver.find_element_by_css_selector(".product-main-info__brand_name")
            brand_name.append(s.text)
            #상품이름
            e=driver.find_element_by_css_selector(".product-main-info__product_name__text")
            product_name.append(e.text) 


        print(i)
        
        

tmp = pd.DataFrame({'product':product_name,'brand':brand_name ,'user' : user_list, 'star':star_list, 'review' : review_list})
# csv 형식으로 저장
tmp.to_csv("data/glowpick_shampoo_real_alllllll5.csv",encoding='utf-8-sig') # 70+153=223
```
